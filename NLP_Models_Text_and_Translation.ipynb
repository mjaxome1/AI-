Hugging Face Examples Models


Translation to French

```python
#Import my libraries
from transformers import pipeline
```

```python
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr") #Model for French Translation

result = translator("Reading and love life of romance", max_length=100, num_return_sequences=1) #Generation length and sequence definition and what exaclty I will be translating

for i, translation in enumerate(result):
    print(f"Translation {i+1}: {translation['translation_text']}") #Print out of the translation and Generated a loop
```

Translation to English



```python
# Define the translator pipeline with the model from Hugging Face
translator = pipeline("translation_es_to_en", model="Helsinki-NLP/opus-mt-es-en")

# Translate the given text
resultado = translator("Lectura y amor por la vida romántica a forever story", max_length=100, num_return_sequences=1)

# Print out the translations
for i, traduccion in enumerate(resultado):
    print(f"Traducción {i+1}: {traduccion['translation_text']}")  # Print out the translation and generate a loop
```

Generator Text Model


```python
generator = pipeline("text-generation", model="distilgpt2") #Model Use distilgpt2 for trained

result = generator("In this course, we will teach you how to", max_length=50, num_return_sequences=1) #the starting of my text In thi course...

for i, generated_text in enumerate(result):
    print(f"Generated text {i+1}: {generated_text['generated_text']}") #Print out my result and create a loop of generated text

```

```python
# Load the text classification pipeline
classifier = pipeline("sentiment-analysis")

# Example texts for classification
texts = [
    "I hate it",
    "This product is terrible, I would never buy it again.",
    "The weather today is amazing.",
    "I'm feeling really sad and frustrated right now."
]

# Classify each text
for text in texts:
    result = classifier(text)
    label = result[0]['label']
    score = result[0]['score']
    print(f"Text: '{text}'")
    print(f"Predicted label: {label}, Score: {score:.4f}")
    print()

```
